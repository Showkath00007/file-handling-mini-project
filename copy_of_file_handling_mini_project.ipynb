```python
File = "students.txt"
def add_student():
  with open(File,"a") as f:
    roll = input("enter the  roll no: ")
    name = input("enter the name of the student: ")
    marks = input("enter the marks: ")
    f.write(f"{roll},{name},{marks}\n")
  print("the student detailes added successfully")
def student_view():
  try:
    with open(File,"r") as f:
      print("the data is found")
      print(f.read())
  except FileNotFoundError:
    print("the data is not found")
def search_student():
  roll = input("enter the roll no: ")
  found = False
  with open(File,"r") as f:
    for line in f:
      if line.startswith(roll + ","):
        print(line)
        found = True
  if not found:
    print("student detailes not found\n")
def delete_student():
  roll = input("enter the roll no: ")
  with open(File,"r") as f:
    lines = f.readlines()
  with open(File,"w") as f:
    for line in lines:
      if not line.startswith(roll+","):
        f.write(line)
  print("student detalils deleted sucussfully\n")
while True:
  print("1.add student")
  print("2.view student")
  print("3.search for student")
  print("4.delete student")
  print("5.exit")
  choice = input("enter the choice: ")
  if choice=="1":
    add_student()
  elif choice=="2":
    student_view()
  elif choice=="3":
    search_student()
  elif choice=="4":
    delete_student()
  elif choice=="5":
    print("exiting program")
    break
  else:
    print("invalid choice")
```
